```python
#                 ------------------ Union of sets -------------------------

#We use the | operator or the union() method to perform the set union operation. For example,

A = {1, 3, 5}
B = {0, 2, 4}
# perform union operation using |
print('Union using |:', A | B)
# perform union operation using union()
print('Union using union():', A.union(B))

# union of two sets

set1 = {"green", "blue"}
setc2 = {"blue", "yellow"}
print("Original sets:")
print(set1)
print(set2)
setc = set1.union(set2)
print("\nUnion of above sets:")
print(setc)
setn1 = {1, 1, 2, 3, 4, 5}
setn2 = {1, 5, 6, 7, 8, 9}
print("\nOriginal sets:")
print(setn1)
print(setn2)
print("\nUnion of above sets:")
setn = setn1.union(setn2)
print(setn)

#  union or more than two sets:

setn1 = {1, 1, 2, 3, 4, 5}
setn2 = {1, 5, 6, 7, 8, 9}
setn3 = {3, 4, 5, 3, 9, 10}
setn4 = {4, 7, 9, 10, 12, 14}

print("\nOriginal sets:")
print(setn1)
print(setn2)
print(setn3)
print(setn4)
print("\nUnion of first three sets:")
setn = setn1.union(setn2, setn3)
print(setn)
print("\nUnion of above four sets:")
setn = setn1.union(setn2, setn3, setn4)
print(setn)

```

```python
#                            ------------------- Intersection  ---------------

# method of intersection

A = {1, 3, 5}
B = {1, 2, 3}

# perform intersection operation using &
print('Intersection using &:', A & B)
# perform intersection operation using intersection()
print('Intersection using intersection():', A.intersection(B)) 

#    taking intersection of two sets

setx = {"green", "blue"}
sety = {"blue", "yellow"}
print("Original set elements:")
print(setx)
print(sety)
print("\nIntersection of two said sets:")
setz = setx & sety
print(setz)

#Compare 3 sets, and return a set with items that is present in all 3 sets:

x = {"a", "b", "c"}
y = {"c", "d", "e"}
z = {"f", "g", "c"}

result = x.intersection(y, z)

print("Intesection of 3 sets is:",result)

```

```python
#                        -------------  Difference of sets --------------------

#  Difference between two sets and syntax

x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

z = x.difference(y) 

print(z)

#    taking difference of two sets

setx = {"green", "blue"}
sety = {"blue", "yellow"}
print("Original set elements:")
print(setx)
print(sety)
print("\ndifference of two said sets:")
setz = setx.difference(sety)
print(setz)

#Compare 3 sets, and return a set with items that is not present in all 3 sets:

x = {"a", "b", "c"}
y = {"c", "d", "e"}
z = {"f", "g", "c"}

result = x.difference(y, z)

print("difference of 3 sets is:",result)
```

```python
#                        -------------  Symmetric_Difference of sets --------------------

#Return a set that contains all items from both sets, except items that are present in both sets:

x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

z = x.symmetric_difference(y)

print(z)

# returns all items to result variable except the items on intersection

A = {'a', 'b', 'c', 'd'}
B = {'c', 'd', 'e' }
result = A.symmetric_difference(B)
print(result)

# returns the symmetric difference of A and B to result variable

A = {'Python', 'Java', 'Go'}
B = {'Python', 'JavaScript', 'C' }
result = A.symmetric_difference(B)

print(result)

# works as (A).symmetric_difference(B)

A = {'a', 'b', 'c', 'd'}
B = {'c', 'd', 'e' }
C = {'i'}

print(A ^ B)

#The symmetric_difference_update() method finds the symmetric difference of two sets (non-similar items of both sets) and updates it to the set that calls the method.

A = {'a', 'c', 'd'}
B = {'c', 'd', 'e' }

# updates A with the symmetric difference of A and B
A.symmetric_difference_update(B)

# symmetric difference of 3 sets
print(A ^ B ^ C)

```
